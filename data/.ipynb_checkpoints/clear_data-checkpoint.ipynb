```python
data = pd.read_csv("data.csv")
header = data.loc[0]
data = data.drop([0])
data.columns = header
data.index = range(0, data.shape[0])
for col in data.columns:
    data[col] = data[col].astype(float)
data
```

```python
import random

def train_test_split(ids, X, Y, rate_test):
    quantidadeIds = len(data['id'].unique())
    quantidadeTeste = int(rate_test * quantidadeIds)
    idsTreino = []
    idsTeste = []
    for i in range(0, quantidadeTeste):
        idEscolhido = random.choice(ids)
        if(idEscolhido not in idsTeste):
            idsTeste.append(idEscolhido)
    for i in range(0, quantidadeIds):
        if(ids[i] not in idsTeste):
            idsTreino.append(ids[i])
    print(idsTeste)
    print(idsTreino)
```
